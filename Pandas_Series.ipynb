## Pandas GroupBy

The groupby()function in pandas is important for data analysis as it allows us to group data by one or more ategories and then apply different functions to those groups .This technique is used for handling large datasets efficiently and performing opereation like aggregation transformation and filteration on grouped data.s me column.

## The groupby() operation is divided into three main step:
- step1: spiltting data into groups the spliting rocess referes to diving the dataset into group based on a particuler condition or key.
  this can be done using the groupby()function by passing one or more columns as keys.

## Methods for spiliting data are as follows:
  group data by a single key :in order to group data with one key we pass only one key as an argument in groupby function. Here we will group the data by the name column

```python
import pandas as pd
```

```python
data1={'Name':['jai','anuj','neha','nitin'],
      'Age':[23,44,56,78],
      'Address':['khairna','nanital','haldwani','ranikhet'],
      'qualification':['Msc','bsc','btech','mtech']}
df=pd.DataFrame(data1)
print(df)
df.groupby('Name')
print(df.groupby('Name').groups)
```
