## Merging DataFrame

Merging DataFrames in Pandas is similar to performing SQL joins. It is useful when we need to combine two DataFrames based on a common column or index. The merge() function provides flexibility for different types of joins.

***There are four basic ways to handle the join (inner, left, right and outer) depending on which rows must retain their data.***

```python
import pandas as pd

data1 = {
    'key': ['K0', 'K1', 'K2', 'K3'],
    'Name': ['Raj', 'Prince', 'Gaurav', 'Anuj'],
    'Age': [27, 24, 22, 32]
}

data2 = {
    'key': ['K0', 'K1', 'K2', 'K3'],
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannauj'],
    'Qualification': ['BTech', 'BA', 'BCom', 'B.hons']
}

df_data1 = pd.DataFrame(data1)
df1_data2 = pd.DataFrame(data2)

print(df_data1, "\n", df1_data2)
```

```python
res=pd.merge(df_data1,df1_data2,on='key')
res
```

```python
res=pd.merge(df_data1,df1_data2,on='Name')
res
```
